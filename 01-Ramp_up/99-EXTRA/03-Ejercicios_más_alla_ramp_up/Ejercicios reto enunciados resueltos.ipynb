```python
# La criba de Erastótenes... SÍ, el mismo que midió la tierra con la sombra de un palo
# A Erastótenes también se le ocurrió un algoritmo para identificar números primos:
# Partiendo de que el 2 es primo, analiza el número 3 y como no se puede dividir con resto cero 
# con ningún número primo menor que el propio 3, significa que 3 es primo.
# Para el 4 tenemos dos números primos [2, 3], como es divisible por uno de ellos (2), no es primo
# Para el 5 tenemos dos números primos [2, 3], como no es divisible por ninguno de ellos, es primo
# Para el 6 tenemos tres números primos [2, 3, 5], como  es divisible por alguno de ellos, no es primo
# [ ... ] 
# Implementa la criba de Erastótenes para los número del 1 al 1000

def erastotenes():
    primos = [2]
    for i in range(2, 1001):
        es_primo = True
        for j in primos:
            if i % j == 0:
                es_primo = False
                break
        if es_primo: 
            primos.append(i)
    return primos

print(erastotenes())
```

```python
# En la actualidad las computadores son capaces de gestionar números tan grandes que es poco probable que manejes una cifra que desborde a tu ordenador
# Sin embargo, esto no siempre fue así y había que buscar otras fórmulas
# Imagínate que tu ordenador solo admite enteros entre -999 y 999. 
# Pero queremos hacer sumas, restas, multiplicaciones y divisiones que superen dicha cantidad. En enteros, de momento.
# La solución consiste en usar dos variables limitadas: parte1 y parte2. Cada vez que superamos 999 en parte1, nos movemos a parte2
# Ahora trabajaremos con listas/tuplas/diccionarios, lo que quieras, con dos elementos, parte1 y parte2
# Ejemplo: el número 245 será (0, 245) o [0, 245] o {"parte2": 0, "parte1": 245}
# Ejemplo: el número 22330 será (22, 339) o [22, 339] o {"parte2": 22, "parte1": 339}
# Pues implementaremos las funciones suma_double(arg1, arg2), resta_double(arg1, arg2), mul_double(arg1, arg2) y div_double(arg1, arg2) 
# arg1 será tu tupla/lista/dict para el primer operador, y arg2 será tu tupla/lista/dict para el segundo operador
# Ejemplo: el número 22330 será (22, 339) o [22, 339] o {"parte2": 22, "parte1": 339} sumado a el número 17930 será (17, 930 ) o [17, 930 ] o {"parte2": 17, "parte1": 930}
# Será: 40260 (40, 260) o [40, 260] o {"parte2": 40, "parte1": 260} 
# No puedes usar +, -, *, //
# Solo puede usar suma(), resta(), mul() y div(), que están preparados para fallar si alguna vez usas un número mayor que 999 o menor que -999

def suma(a, b, limit = 999):
     if (abs(a) > limit) or (abs(b) > limit) or (abs(a + b) > limit):
        raise Exception("Has superado", limit)
     else:
         return a + b
     
def resta(a, b, limit = 999):
     if (abs(a) > limit) or (abs(b) > limit) or (abs(a - b) > limit):
        raise Exception("Has superado", limit)
     else:
         return a - b

def mul(a, b, limit = 999):
     if (abs(a) > limit) or (abs(b) > limit) or (abs(a * b) > limit):
        raise Exception("Has superado", limit)
     else:
         return a * b
     
def div(a, b, limit = 999):
     if (abs(a) > limit) or (abs(b) > limit) or (abs(a // b) > limit):
        raise Exception("Has superado", limit)
     else:
         return a // b
```

```python
# Esto lo hice hace unos 20 años y es más dificil de lo que creía recordar!!
# Hay que analizar despacio todas las posibilidades:
# Suma de dos positivos sin superar XXX999 
# Suma de dos negativos sin superar XXX999
# Suma de dos números con distinto signo: es una resta
# Resta de dos números positivos sin alcanzar XXX000
# Resta de dos números negativos sin alcanzar XXX000
# Resta de dos números con distingo signo: es una suma
# No he hecho la multiplicación o división. Pero hay una forma fácil sin hacer apenas nada más!

# Funciones auxiliares
 

def zero(num):
    # El número es cero
    return num[1] == 0 and num[0] == 0

def signo_pos (num):
    # El número es positivo o cero
    op1 = num[0] >= 0
    op2 = num[1] >= 0
    return op1 and op2

def mismo_signo(a,b):
    # Los números tienen el mismo signo
    if zero(a) or zero(b):
        return True
    op1 = (signo_pos(a) and signo_pos(b))  # O son positivos/mayor que cero
    op2 = not signo_pos(a) and not signo_pos(b) # O son negativos
    return op1 or op2


def max_suma(num, limit = 999): 
      # ¿Cuanto es lo máximo que puedo sumar a un número antes de superar 999?
      return resta(limit, num)

def max_resta(num, limit = 999): 
      # ¿Cuanto es lo máximo que puedo resta a un número antes de superar -999?
      return resta(-limit, num)


def inversa(a,b):
    # Si sumo/resto dos números negativos, puedo hacer una suma/resta "normal" 
    # Pero cambio el signo al final
    if (not signo_pos(a) and (not signo_pos(b))):
        return -1
    else:
        return 1
    

def cambio_signo(num):
    # Cambiar el signo a un número
    return[-num[0], -num[1]]

def reordenar_resta(a, b):
    # En la resta pongo siempre el número positivo primero y el negativo después
    if signo_pos(a):
        return [a,b]
    else:
        return [b,a]

def resta_signo_resultado(a, b):
    # Calculo el signo de la resta. 
    # Si el primer operando es mayor, conservará el signo
    # Si el primer operando es menor, es una resta que debería cambiar de signo
    if abs(a[0]) > abs(b[0]):
        return  True
    elif abs(a[0]) < abs(b[0]):
        return False
    elif (a[0] == b[0]) and (abs(a[1]) >= abs(b[1])):
        return True
    else:
        return False
    
def lastre(num1, num2):
    # En una resta, si la segunda parte del que va restando es mayor,
    # Arrastramos 1 que hay que quitar de la primera parte del número
    if num1 >= num2:
        return 0
    else:
        return 1
    
def resta_primer_operando(num1, num2, n_lastre):
    # La resta de la primera parte del número
    # Una resta normal
    # Quizá quitar el lastre
    resta1 = resta(num1, num2)
    resta2 = resta(resta1, n_lastre)
    return resta2

def resta_segundo_operando(num1, num2, n_lastre, limit = 999):
     # La resta de la segunad parte del número
     # Si genero lastre, habría que sumar el límite + 1
     # Ejemplo 15304 - 13502 es 1802. 304 - 502 son -198. -198 + 999 son 801. 801 + 1 son 802 
     resta1 = resta(num1, num2)
     resta2 = suma(resta1, n_lastre*limit)
     resta3 = suma(resta2, n_lastre)
     return resta3


def transform_to_num(l):
    # transformar una lista de 2 números menores de 999 en el número tradicional
    string = int(str(l[0]))*1000 + int(str(l[1]))
    return  int(string)     

def check_resta(l1, l2):
    # Para chequear que hacemos bien la resta
    num1 = transform_to_num(l1)
    num2 = transform_to_num(l2)
    return num1 - num2

def check_suma(l1, l2):
    # Para chequear que hacemos bien la suma
    num1 = transform_to_num(l1)
    num2 = transform_to_num(l2)
    return num1 + num2

def operacion(oper, a, b):
   # Realizo la operación asignada
   signo_igual = mismo_signo(a,b) # Compruebo si son del mismo signo
   inv = inversa(a, b) # Me acuerdo si son ambos negativos. Si son ambos negativos, hago la operación como si fueran negativos e invierto el valor resultante
   if (oper == "suma" and signo_igual): # Si es suma y son del mismo signo. Una suma normal, cambio de signo si eran ambos negativos
      #print("Suma normal")
      return suma_compleja(a, b, inv) 
   elif (oper == "suma" and not signo_igual): # Si es suma y son de distinto signo. Es una resta, pongo el número positivo primero, cambio el signo al segundo y hago resta normal
      #print("Suma cuando los números tienen distinto signo, es una resta")
      a, b = reordenar_resta(a, b) # Ponemos el positivo primero
      b = cambio_signo(b) # Cambiamos de signo al segundo 
      return resta_compleja(a, b) # Resta
   elif (oper == "resta" and signo_igual): # Resta del mismo signo es resta noraml
       #print("Resta normal")
       return resta_compleja(a, b)
   elif (oper == "resta" and not signo_igual): # Resta cuando los números tienen distinto signo, es una suma
       #print("Resta cuando los números tienen distinto signo, es una suma")
       primero_pos = not signo_pos(a)  # Guardo si el primer numero era positivo
       a, b = reordenar_resta(a, b) # Ponemos el positivo primero
       a = cambio_signo(a) # Cambiamos de signo al positivo 
       inv = inversa(a, b) # Calculamos si hay que cambiar de signo al final
       resta =  suma_compleja(a, b, inv) # La resta se transforma en suma
    
       if primero_pos: # Si el primero era positivo, el resultado esta bien
           return resta
       else: # Si era negativo, como hemos invertido la operación, hay que cambiar el signo
           return cambio_signo(resta)
           
     
def suma_compleja(a, b, inv=1, limit = 999):
    # La suma con números de dos partes
    if zero(a) or zero(b): # Si alguno es cero, nada cambia
        return [suma(a[0], b[0]), suma(a[1], b[1])]
    if (inv == -1): # Si estamos sumando dos negativos, cambiamos los signos
        a = cambio_signo(a)
        b = cambio_signo(b)
    max_suma_numero = max_suma(a[1]) # Calculamos el maximo número a sumar hasta llegar a 999
    if max_suma_numero >= b[1]: # Si no nos pasamos de 999
        return (suma(a[0], b[0])*inv, suma(a[1], b[1])*inv) # Nada raro
    else: # Si nos pasamos de 999
        primer_digito = resta(resta(a[1], resta(limit, b[1])), 1) # Hay que dar la vuelta a 999. Ejemplo 15802 + 12203. 28005. 802 + 203 = 1005 luego 802 -796 son 6 - 1 son 5
        return (suma(suma(a[0], b[0])*inv, 1*inv), primer_digito*inv)


def resta_compleja(a, b, limit = 999):
     # La suma con números de dos partes
    if zero(a) or zero(b):  # Si alguno es cero, nada cambia
        return [resta(a[0], b[0]), resta(a[1], b[1])]
    if resta_signo_resultado(a, b): # Si al restar conservamos el signo del primero:
        n_lastre = lastre(a[1], b[1]) # Calculamos si tenemos que restar 1 a la primera parta
        resta_1 = resta_primer_operando(a[0], b[0], n_lastre) # Restamos el primer operando
        resta_2 = resta_segundo_operando(a[1], b[1], n_lastre, limit) # Restamos el segundo operando
        return [resta_1, resta_2]
    else: # Si no conservamos el signo del primero, es una resta cambiada de signo
        return cambio_signo(resta_compleja(b, a, limit)) 
```
